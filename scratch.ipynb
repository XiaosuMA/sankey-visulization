```python
def plot_normal_distributions():
    x = np.linspace(0, 100, 1000)
    y1 = (1 / (10 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - 50) / 10) ** 2)
    y2 = (1 / (20 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - 50) / 20) ** 2)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y1, label='µ = 50, $\sigma$ = 10', linestyle='--', color='gray')
    plt.plot(x, y2, label='µ = 50, $\sigma$ = 20', linestyle='-', color='black')
    plt.title('')
    plt.xlabel('Quantity')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(False)
    plt.show()

if __name__ == "__main__":
    plot_normal_distributions()
```

```python
# plotting two different normal distributions for newvendor model with mean of 50 in one plot, one drawm with std of 10 and the other with std of 20
# the one with std of 10 should be draw with dashed line adn the other wth solid line
# remove the gride lines and add a legend to the plot
# the area under the curve between 0 to 95% of the total area should be shaded in gray and mark the the quantity of reach distribution at 95% of the total area

def plot_normal_distributions():
    x = np.linspace(0, 100, 1000)
    y1 = (1 / (10 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - 50) / 10) ** 2)
    y2 = (1 / (20 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - 50) / 20) ** 2)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y1, label='µ = 50, $\sigma$ = 10', linestyle='--', color='gray')
    plt.plot(x, y2, label='µ = 50, $\sigma$ = 20', linestyle='-', color='black')

    # Calculate exact 95th percentile using scipy.stats
    x_95_y1 = stats.norm.ppf(0.95, loc=50, scale=10)  # 95th percentile for N(50, 10)
    x_95_y2 = stats.norm.ppf(0.95, loc=50, scale=20)  # 95th percentile for N(50, 20)
    
    # Shading the area under the curve for 95% of the total area
    plt.fill_between(x, y1, where=(x <= x_95_y1), color='green', alpha=0.2)
    plt.fill_between(x, y2, where=(x <= x_95_y2), color='gray', alpha=0.5)
    
    # Mark the 95th percentile points
    # add a arrow from first vertical line towar the second vertical line
    plt.axvline(x=x_95_y1, color='green', linestyle=':', alpha=0.3)
    plt.axvline(x=x_95_y2, color='black', linestyle=':', alpha=0.5)
    plt.annotate('', xy=(x_95_y2, 0.031), xytext=(x_95_y1, 0.031),
                 arrowprops=dict(arrowstyle='->', color='gray', lw=1.5))
    
    # Add text annotations for the percentile values
    # text should be place at the right of the vertical line
    plt.text(x_95_y1+2, 0.035, f'$\sigma$=10\nStock {x_95_y1:.1f}', 
             horizontalalignment='left', fontsize=10)
    plt.text(x_95_y2+2, 0.035, f'$\sigma$=20\nStock {x_95_y2:.1f}', 
             horizontalalignment='left', fontsize=10)

    plt.title('')
    plt.xlabel('Quantity')
    plt.ylabel('Probability Density')
    plt.xlim(0, 100)
    plt.ylim(0, 0.05)
    plt.legend()
    plt.grid(False)
  

if __name__ == "__main__":
    plot_normal_distributions()
    plt.savefig('normal_distributions.png', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
import plotly.graph_objects as go
# Remove numpy import as it's not actually needed for this visualization
# import numpy as np

def create_modal_shift_sankey():
    """
    Create a Sankey diagram with custom layer names
    """
    
    # Define nodes with custom layer structure
    node_labels = [
        # First layer: Geographically Contingents (left side)
        "Truck-Based",
        "Light CV", 
        
        # Second layer: Challenges (middle)
        "Truck-based",
        "Light CV.",
        "Micromobility",
        
        # Third layer: Concepts (right side)
        "truck-based",
        "light cv",
        "suppressed", 
        "micromobility",
        "surface robots",
        "drones"
    ]
    
    # Define node colors
    node_colors = [
        # Source nodes
        "#9E9E9E",  # Truck-Based (gray)
        "#BBDEFB",  # Light CV (light blue)
        
        # Intermediate nodes  
        "#757575",  # Truck-based (dark gray)
        "#90CAF9",  # Light CV. (blue)
        "#A5D6A7",  # Micromobility (green)
        
        # Target nodes
        "#616161",  # truck-based (dark gray)
        "#64B5F6",  # light cv (blue)
        "#9E9E9E",  # suppressed (gray)
        "#81C784",  # micromobility (green)
        "#F8BBD9",  # surface robots (pink)
        "#4DD0E1"   # drones (cyan)
    ]
    
    # Define flows (source, target, value)
    # Format: [source_index, target_index, flow_value]
    flows = [
        # From Truck-Based (0) to intermediate
        [0, 2, 40],  # to Truck-based
        [0, 3, 15],  # to Light CV.
        [0, 4, 5],   # to Micromobility
        
        # From Light CV (1) to intermediate  
        [1, 2, 5],   # to Truck-based
        [1, 3, 35],  # to Light CV.
        [1, 4, 15],  # to Micromobility
        
        # From intermediate to final destinations
        # From Truck-based (2)
        [2, 5, 35],  # to truck-based
        [2, 6, 8],   # to light cv
        [2, 7, 2],   # to suppressed
        
        # From Light CV. (3)
        [3, 6, 25],  # to light cv
        [3, 7, 5],   # to suppressed
        [3, 8, 15],  # to micromobility
        [3, 10, 5],  # to drones
        
        # From Micromobility (4)
        [4, 8, 15],  # to micromobility
        [4, 9, 3],   # to surface robots
        [4, 10, 2],  # to drones
    ]
    
    # Create the Sankey diagram with layer annotations
    fig = go.Figure(data=[go.Sankey(
        node = dict(
            pad = 15,
            thickness = 20,
            line = dict(color = "black", width = 0.5),
            label = node_labels,
            color = node_colors,
            x = [0.01, 0.01,  # Geographically Contingents layer
                 0.5, 0.5, 0.5,  # Challenges layer
                 0.99, 0.99, 0.99, 0.99, 0.99, 0.99],  # Concepts layer
            y = [0.8, 0.2,  # Geographically Contingents positions
                 0.9, 0.5, 0.1,  # Challenges positions
                 0.95, 0.75, 0.55, 0.35, 0.15, 0.05]  # Concepts positions
        ),
        link = dict(
            source = [flow[0] for flow in flows],
            target = [flow[1] for flow in flows], 
            value = [flow[2] for flow in flows],
            color = ["rgba(158,158,158,0.5)" if flow[0] < 2 else 
                    "rgba(187,222,251,0.5)" if flow[1] in [6,7] else
                    "rgba(165,214,167,0.5)" if flow[1] in [8,9] else  
                    "rgba(77,208,225,0.5)" if flow[1] == 10 else
                    "rgba(200,200,200,0.5)" for flow in flows]
        )
    )])

    fig.update_layout(
        title={
            'text': "Figure 12. Modelled modal shift with increased last-metre alternatives (trips)",
            'x': 0.02,
            'y': 0.95,
            'xanchor': 'left',
            'font': {'size': 16}
        },
        font_size=12,
        width=900,
        height=650,  # Increased height to accommodate layer labels
        margin=dict(l=50, r=50, t=100, b=50),  # Increased top margin for labels
        annotations=[
            # Layer labels
            dict(x=0.01, y=1.08, text="<b>Geographically Contingents</b>", 
                 showarrow=False, font=dict(size=14, color="black"), 
                 xanchor="left", xref="paper", yref="paper"),
            dict(x=0.5, y=1.08, text="<b>Challenges</b>", 
                 showarrow=False, font=dict(size=14, color="black"), 
                 xanchor="center", xref="paper", yref="paper"),
            dict(x=0.99, y=1.08, text="<b>Concepts</b>", 
                 showarrow=False, font=dict(size=14, color="black"), 
                 xanchor="right", xref="paper", yref="paper")
        ]
    )
    
    return fig

# Create and display the diagram
fig = create_modal_shift_sankey()
fig.show()
```

```python

# EXPORT FUNCTIONS FOR POWERPOINT INTEGRATION
def export_sankey_for_ppt(fig, output_path, filename="sankey_diagram"):
    """
    Export Sankey diagram in multiple formats for PowerPoint integration
    
    Parameters:
    fig: plotly figure object
    output_path: str, directory path to save files
    filename: str, base filename (without extension)
    """
    import os
    
    # Create output directory if it doesn't exist
    os.makedirs(output_path, exist_ok=True)
    
    # 1. Export as interactive HTML (can be embedded in PPT)
    html_file = os.path.join(output_path, f"{filename}.html")
    fig.write_html(html_file, 
                   include_plotlyjs='cdn',  # Use CDN for smaller file size
                   config={'displayModeBar': False})  # Hide toolbar for cleaner look
    print(f"Interactive HTML saved: {html_file}")
    
    # 2. Export as standalone HTML (no internet required)
    html_standalone = os.path.join(output_path, f"{filename}_standalone.html")
    fig.write_html(html_standalone, 
                   include_plotlyjs=True,  # Include plotly.js inline
                   config={'displayModeBar': False})
    print(f"Standalone HTML saved: {html_standalone}")
    
    # 3. Export as static PNG (for fallback or thumbnails)
    png_file = os.path.join(output_path, f"{filename}.png")
    try:
        fig.write_image(png_file, width=1200, height=800, scale=2)
        print(f"PNG image saved: {png_file}")
    except Exception as e:
        print(f"PNG export failed (install kaleido: pip install kaleido): {e}")
    
    # 4. Export as SVG (vector format, good for scaling)
    svg_file = os.path.join(output_path, f"{filename}.svg")
    try:
        fig.write_image(svg_file, width=1200, height=800)
        print(f"SVG image saved: {svg_file}")
    except Exception as e:
        print(f"SVG export failed: {e}")
    
    # 5. Create PowerPoint-optimized HTML with custom styling
    ppt_html_file = os.path.join(output_path, f"{filename}_ppt_optimized.html")
    create_ppt_optimized_html(fig, ppt_html_file)
    print(f"PPT-optimized HTML saved: {ppt_html_file}")
    
    return {
        'html': html_file,
        'html_standalone': html_standalone,
        'png': png_file,
        'svg': svg_file,
        'ppt_html': ppt_html_file
    }

def create_ppt_optimized_html(fig, output_path):
    """
    Create PowerPoint-optimized HTML with custom styling
    """
    # Configure figure for PowerPoint
    fig_copy = fig
    fig_copy.update_layout(
        margin=dict(l=20, r=20, t=40, b=20),
        paper_bgcolor='white',
        plot_bgcolor='white',
        font=dict(size=14),
        title=dict(font=dict(size=16, color='black'))
    )
    
    # Custom HTML template optimized for PowerPoint
    html_template = """
    <!DOCTYPE html>
    <html>
    <head>
        <meta charset="utf-8">
        <title>Sankey Diagram for PowerPoint</title>
        <script src="https://cdn.plot.ly/plotly-latest.min.js"></script>
        <style>
            body {{
                margin: 0;
                padding: 10px;
                font-family: Arial, sans-serif;
                background-color: white;
            }}
            #plotly-div {{
                width: 100%;
                height: 90vh;
            }}
        </style>
    </head>
    <body>
        <div id="plotly-div"></div>
        <script>
            var plotData = {plot_data};
            var layout = {layout};
            var config = {{
                displayModeBar: false,
                responsive: true,
                displaylogo: false
            }};
            
            Plotly.newPlot('plotly-div', plotData, layout, config);
            
            // Resize handler for PowerPoint
            window.addEventListener('resize', function() {{
                Plotly.Plots.resize('plotly-div');
            }});
        </script>
    </body>
    </html>
    """
    
    # Generate the HTML
    plot_json = fig_copy.to_json()
    import json
    plot_dict = json.loads(plot_json)
    
    html_content = html_template.format(
        plot_data=json.dumps(plot_dict['data']),
        layout=json.dumps(plot_dict['layout'])
    )
    
    with open(output_path, 'w', encoding='utf-8') as f:
        f.write(html_content)

# USAGE EXAMPLE
def main_export_example():
    """
    Complete example of creating and exporting Sankey diagram
    """
    # Create the diagram
    fig = create_modal_shift_sankey()
    
    # Define your output path
    output_directory = "C:/Users/YourName/Desktop/sankey_exports"  # Change this path
    
    # Export in multiple formats
    exported_files = export_sankey_for_ppt(
        fig=fig,
        output_path=output_directory,
        filename="modal_shift_sankey"
    )
    
    print("\n=== EXPORT COMPLETE ===")
    print("Files created:")
    for format_type, filepath in exported_files.items():
        print(f"  {format_type}: {filepath}")
    
    print("\n=== POWERPOINT INTEGRATION METHODS ===")
    print("1. WEB OBJECT METHOD:")
    print("   - Insert > Get Add-ins > Web Viewer")
    print(f"   - Use file: {exported_files['html_standalone']}")
    
    print("\n2. HYPERLINK METHOD:")
    print("   - Insert > Link > Link to File")
    print(f"   - Link to: {exported_files['html']}")
    
    print("\n3. STATIC IMAGE METHOD:")
    print("   - Insert > Pictures > From File")
    print(f"   - Use file: {exported_files.get('png', 'PNG export failed')}")
    
    return exported_files

# Quick export function with minimal setup
def quick_export(output_path="./sankey_export", filename="sankey_diagram"):
    """
    Quick export function - just specify path and filename
    """
    fig = create_modal_shift_sankey()
    return export_sankey_for_ppt(fig, output_path, filename)

# Uncomment to run the export:
exported = main_export_example()

# Or use quick export:
# quick_export("C:/path/to/your/folder", "my_sankey_diagram")
```

```python

# EXPORT FUNCTIONS FOR POWERPOINT INTEGRATION
def export_sankey_for_ppt(fig, output_path, filename="sankey_diagram"):
    """
    Export Sankey diagram in multiple formats for PowerPoint integration
    
    Parameters:
    fig: plotly figure object
    output_path: str, directory path to save files
    filename: str, base filename (without extension)
    """
    import os
    
    # Create output directory if it doesn't exist
    os.makedirs(output_path, exist_ok=True)
    
    # 1. Export as interactive HTML (can be embedded in PPT)
    html_file = os.path.join(output_path, f"{filename}.html")
    fig.write_html(html_file, 
                   include_plotlyjs='cdn',  # Use CDN for smaller file size
                   config={'displayModeBar': False})  # Hide toolbar for cleaner look
    print(f"Interactive HTML saved: {html_file}")
    
    # 2. Export as standalone HTML (no internet required)
    html_standalone = os.path.join(output_path, f"{filename}_standalone.html")
    fig.write_html(html_standalone, 
                   include_plotlyjs=True,  # Include plotly.js inline
                   config={'displayModeBar': False})
    print(f"Standalone HTML saved: {html_standalone}")
    
    # 3. Export as static PNG (for fallback or thumbnails)
    png_file = os.path.join(output_path, f"{filename}.png")
    try:
        fig.write_image(png_file, width=1200, height=800, scale=2)
        print(f"PNG image saved: {png_file}")
    except Exception as e:
        print(f"PNG export failed (install kaleido: pip install kaleido): {e}")
    
    # 4. Export as SVG (vector format, good for scaling)
    svg_file = os.path.join(output_path, f"{filename}.svg")
    try:
        fig.write_image(svg_file, width=1200, height=800)
        print(f"SVG image saved: {svg_file}")
    except Exception as e:
        print(f"SVG export failed: {e}")
    
    # 5. Create PowerPoint-optimized HTML with custom styling
    ppt_html_file = os.path.join(output_path, f"{filename}_ppt_optimized.html")
    create_ppt_optimized_html(fig, ppt_html_file)
    print(f"PPT-optimized HTML saved: {ppt_html_file}")
    
    return {
        'html': html_file,
        'html_standalone': html_standalone,
        'png': png_file,
        'svg': svg_file,
        'ppt_html': ppt_html_file
    }

def create_ppt_optimized_html(fig, output_path):
    """
    Create PowerPoint-optimized HTML with custom styling
    """
    # Configure figure for PowerPoint
    fig_copy = fig
    fig_copy.update_layout(
        margin=dict(l=20, r=20, t=40, b=20),
        paper_bgcolor='white',
        plot_bgcolor='white',
        font=dict(size=14),
        title=dict(font=dict(size=16, color='black'))
    )
    
    # Custom HTML template optimized for PowerPoint
    html_template = """
    <!DOCTYPE html>
    <html>
    <head>
        <meta charset="utf-8">
        <title>Sankey Diagram for PowerPoint</title>
        <script src="https://cdn.plot.ly/plotly-latest.min.js"></script>
        <style>
            body {{
                margin: 0;
                padding: 10px;
                font-family: Arial, sans-serif;
                background-color: white;
            }}
            #plotly-div {{
                width: 100%;
                height: 90vh;
            }}
        </style>
    </head>
    <body>
        <div id="plotly-div"></div>
        <script>
            var plotData = {plot_data};
            var layout = {layout};
            var config = {{
                displayModeBar: false,
                responsive: true,
                displaylogo: false
            }};
            
            Plotly.newPlot('plotly-div', plotData, layout, config);
            
            // Resize handler for PowerPoint
            window.addEventListener('resize', function() {{
                Plotly.Plots.resize('plotly-div');
            }});
        </script>
    </body>
    </html>
    """
    
    # Generate the HTML
    plot_json = fig_copy.to_json()
    import json
    plot_dict = json.loads(plot_json)
    
    html_content = html_template.format(
        plot_data=json.dumps(plot_dict['data']),
        layout=json.dumps(plot_dict['layout'])
    )
    
    with open(output_path, 'w', encoding='utf-8') as f:
        f.write(html_content)

# USAGE EXAMPLE
def main_export_example():
    """
    Complete example of creating and exporting Sankey diagram
    """
    # Create the diagram
    fig = create_modal_shift_sankey()
    
    # Define your output path
    output_directory = "/Users/xiaosu/Nextcloud/00 CSA Shared Data/Promotion Xiaosu Li Shared Ressource/0-Current-Work/Illustrations"  # Change this path
    
    # Export in multiple formats
    exported_files = export_sankey_for_ppt(
        fig=fig,
        output_path=output_directory,
        filename="modal_shift_sankey"
    )
    
    print("\n=== EXPORT COMPLETE ===")
    print("Files created:")
    for format_type, filepath in exported_files.items():
        print(f"  {format_type}: {filepath}")
    
    print("\n=== POWERPOINT INTEGRATION METHODS ===")
    print("1. WEB OBJECT METHOD:")
    print("   - Insert > Get Add-ins > Web Viewer")
    print(f"   - Use file: {exported_files['html_standalone']}")
    
    print("\n2. HYPERLINK METHOD:")
    print("   - Insert > Link > Link to File")
    print(f"   - Link to: {exported_files['html']}")
    
    print("\n3. STATIC IMAGE METHOD:")
    print("   - Insert > Pictures > From File")
    print(f"   - Use file: {exported_files.get('png', 'PNG export failed')}")
    
    return exported_files

# Quick export function with minimal setup
def quick_export(output_path="./sankey_export", filename="sankey_diagram"):
    """
    Quick export function - just specify path and filename
    """
    fig = create_modal_shift_sankey()
    return export_sankey_for_ppt(fig, output_path, filename)

# Uncomment to run the export:
exported = main_export_example()

```
